```python
def directing_decryption(x, step_shift):
    """
    Функция для дешифрования текста с известным шагом сдвига
    """
    y = ''
    for i in range(len(x)):
        if x[0] > 'z':  # Русский текст
            if 1040 <= ord(x[i]) <= 1071 and 1040 <= ord(x[i]) - step_shift <= 1071:
                y += chr(ord(x[i]) - step_shift)
            elif 1040 <= ord(x[i]) <= 1071 and ord(x[i]) - step_shift < 1040:
                y += chr(1072 - (step_shift - (ord(x[i]) - 1040)))
            elif 1072 <= ord(x[i]) <= 1103 and 1072 <= ord(x[i]) - step_shift <= 1103:
                y += chr(ord(x[i]) - step_shift)
            elif 1072 <= ord(x[i]) <= 1103 and ord(x[i]) - step_shift < 1072:
                y += chr(1104 - (step_shift - (ord(x[i]) - 1072)))
            else:
                y += x[i]

```
